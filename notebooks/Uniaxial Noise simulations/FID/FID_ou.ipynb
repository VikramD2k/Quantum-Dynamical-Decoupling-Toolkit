```julia
parameters = get_preset_params("ou")
S_func = ω-> S(ω; parameters...)

# Run FID simulation
T_max = 10.0         # Total maximum evolution time
dt = 1e-1             # Time step (resolution of T values)
n_realizations = 1000 # Number of noise samples
dc = 0.0

T_vals, avg_fids = fast_average_fidelity_vs_time(S_func;
                                                 T_max=T_max,
                                                 dt=dt,
                                                 n_realizations=n_realizations,
                                                 target_std=1.0,dc = dc,
                                                 seed_offset=rand(Int),
                                                 use_gpu=true,
                                                 verbose=true)

# Plot the results
# plot(T_vals, avg_fids,
#      xlabel="Time (T)",
#      ylabel="⟨Fidelity⟩",
#      title="FID Decay under OU Noise",
#      lw=2, marker=:circle,
#      legend=false)
```

```julia
using LsqFit

# Define the fit model: F(t) = 0.5 + 0.5 * exp(-Γ * t^2)
fit_model(t, p) = 0.5 .+ 0.5 .* exp.(-p[1] .* t)

# Perform fit
fit = curve_fit(fit_model, T_vals, avg_fids, [0.001])  # initial guess for Γ

Γ_fit = fit.param[1]
println("Fitted Gamma (Γ): ", Γ_fit)
```

```julia
plot(T_vals, avg_fids,
     label="Simulated ⟨F(t)⟩", lw=2, marker=:circle,
     xlabel="Time (T)", ylabel="⟨Fidelity⟩",
     title="FID Decay under OU Noise")

plot!(T_vals, fit_model(T_vals, fit.param),
      label="Fit: ½ + ½·e^(–Γt),  Γ = $(round(Γ_fit, digits=4))", lw=2, linestyle=:dash)
```

```julia
using Plots
params = get_preset_params("ou")
S_ou = ω -> S(ω; params...)

# Time grid
dt = 1e-2
T_max = 25
τ_vals = 0.0:dt:T_max

# Run using FID
coh_fid_truncated = exp.(- χ(S_ou, filter_fid, τ_vals,dt; truncate=true, parallelize=true))
coh_fid_untruncated = exp.(- χ(S_ou, filter_fid, τ_vals, dt; truncate=false, parallelize=true))
# Plot
plot(τ_vals, coh_fid_truncated, lw=2, label="FID_truncated", xlabel="τ", ylabel="⟨F(τ)⟩", title="Coherence Decay from Filter Function", color =:blue)
plot!(τ_vals, coh_fid_untruncated, lw=2, label="FID_untruncated", color =:red)
```

### To check if the final average fidelity converges as we increase $dt$

```julia
# Run FID simulation
T_max = 1.0         # Total maximum evolution time
n_realizations = 1000 # Number of noise samples
σ = 1.0
```

```julia
dt = 1e-1             # Time step (resolution of T values)

T_vals, avg_fids = fast_average_fidelity_vs_time(S_func;
                                                 T_max=T_max,
                                                 dt=dt,
                                                 n_realizations=n_realizations,
                                                 target_std=σ,
                                                 dc=0.0,
                                                 seed_offset=rand(Int),
                                                 use_gpu=true,
                                                 verbose=true)

# Plot the results
plot(T_vals, avg_fids,
     xlabel="Time (T)",
     ylabel="⟨Fidelity⟩",
     title="FID Decay under White Noise",
     lw=2, marker=:circle,
     legend=false)
# Define the fit model: F(t) = 0.5 + 0.5 * exp(-Γ * t^2)
fit_model(t, p) = 0.5 .+ 0.5 .* exp.(-p[1] .* t)

# Perform fit
fit = curve_fit(fit_model, T_vals, avg_fids, [0.001])  # initial guess for Γ
Γ_fit = fit.param[1]
plot!(T_vals, fit_model(T_vals, fit.param),
      label="Fit: ½ + ½·e^(–Γt),  Γ = $(round(Γ_fit, digits=4))", lw=2, linestyle=:dash)
```

```julia
print("Final average fidelity for dt = $dt: ", avg_fids[end])
```
